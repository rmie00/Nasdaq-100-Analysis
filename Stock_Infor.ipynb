```python
import pandas as pd
import numpy as np
import datetime as dt
import yfinance as yf
```

```python
index = pd.read_csv('tech_index.csv')
```

```python
indexs = (index['Symbol']).unique()
```

```python
### Download Stock Information From Yahoo For Top 1000 Tech Comapnies

start = dt.datetime(2019,12,1)
end = dt.datetime(2023,12,1)


stock_list = indexs.tolist()

stock_data = dict()
for symbol in stock_list:
    try:
        stock_data[symbol] = yf.download(symbol, start=start, end=end)
    except:
        print(f'Error for fetching data for {symbol}')
        continue
```

```python
### Daily Stock Price DataFrame###
data_df = pd.concat(stock_data.values(), keys=stock_data.keys(), axis=0).reset_index()
data_df
```

```python
tech_name = index.set_index('Symbol')['Name'].to_dict()
for index, row in data_df.iterrows():
    if row['level_0'] in tech_name:
        data_df.at[index, 'Name'] = tech_name[row['level_0']]
```

```python
data_df.set_index('Name', inplace=True)
data_df.index
```

```python
data_df.to_csv('Top_948_Tech_Company_Stocks.csv', index=True)
```
