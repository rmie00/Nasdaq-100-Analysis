```python
nasdaq_df = pd.DataFrame.from_dict(nasdaq, orient='index').reset_index()
```

```python
nasdaq_df = nasdaq_df.drop(nasdaq_df.columns[0], axis=1)
nasdaq_df
```

```python
start = dt.datetime(2019,12,1)
end = dt.datetime(2023,12,1)

symbol = nasdaq_df['Symbol'].tolist()
nasdaq_stock = dict()
for stock in symbol:
    try:
        nasdaq_stock[stock] = yf.download(stock, start=start, end=end)
    except:
        print(f'Couldnt Retrieve {stock} data')
        continue
```

```python
nasdaq_stock_df = pd.concat(nasdaq_stock.values(), keys=nasdaq_stock.keys(), axis=0).reset_index()
nasdaq_stock_df
stock_name_dict = nasdaq_df.set_index('Symbol')['index'].to_dict()
for index, row in nasdaq_stock_df.iterrows():
    if row['level_0'] in stock_name_dict:
        nasdaq_stock_df.at[index, 'Name'] = stock_name_dict[row['level_0']]
```

```python
nasdaq_stock_df.rename(columns ={'level_0':'Symbol'}, inplace=True)
```
